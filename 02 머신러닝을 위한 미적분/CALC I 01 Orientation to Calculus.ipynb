## 1.2 A Brief History
- ~1800 BCE: Egyption papyrus with area calculations
- "Method of exhaustion" developed by:
  - Greeks Eudoxus (~400 BCE) and Archimedes (~250 BCE)
  - Independently by Chinese Liu Hui (~250 CE)
- Arab al-Haytham (11th c. CE) used integrals to calculate volumns
- Indians (14th c.) had differentiation-like methods
- 17th c.: $Gottfried$ $Leibniz$ (Germany) & $Isaac$ $Newton$ (English)
  - Independently developed modern calculus
  - Includes higher-order differentation and integration
  - Includes rules like **roduct rule** & **chain rule** (Calc I)
  - $Leibniz$ named it and devised the notation 
  - $Newton$ applied it first to physics: 
    - Laws of motion
    - Gravity

## 1.3 Calculus of the Infinitesimals

As integral acuracy improves as we approach an ifinite-sided polygon, so too does **differential accuracy** improve as we approach a curve infinitely closely. 


```python
x = np.linspace(-10, 10, 10000)   # start, finish, n points  # space being 0.002 between points
```

if $y = x^2 + 2x + 2$:

```python
y = x**2 + 2*x + 2
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, c='dodgerblue')
```

-   There are no straight lines on the curve.
-   If we zoom in infinitely close, however, we observe curves that approach lines.
-   This enables us to find a slope $m$ (tangent) anywhere on the curve, including to identify where $m = 0$: (where slope is flat.)

A quick recap of slopes:   
slope.gif  
image from [Lumen, Module 15: Graphs "Interpreting Slope in Equations and Graphs"](https://courses.lumenlearning.com/wm-developmentalemporium/chapter/read-or-watch-house-value/)


```python
fig, ax = plt.subplots(1, 5, figsize=(30, 5))

x_scalers = [(-2, 0), (-1.5, -0.5), (-1.1, -0.9), (-1.1, -0.9), (-1.01, -0.99)]
y_scalers = [(0, 2), (0.5, 1.5), (0.9, 1.1), (0.9, 1.1), (0.99, 1.01)]

for i in range(5):
  ax[i].set_xlim(x_scalers[i])
  ax[i].set_ylim(y_scalers[i])
  ax[i].plot(x,y)
fig.suptitle("approaching slope 0", fontsize=20, fontweight="bold")
plt.show()
```
