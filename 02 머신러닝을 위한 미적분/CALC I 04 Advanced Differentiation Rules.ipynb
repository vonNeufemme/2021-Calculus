# 4. Advanced Differentiation Rules

````python
import numpy as np
import matplotlib.pyplot as plt

# plot one graph
def plot_graph(x, y, xlimits=None, ylimits=None):
  fig, ax = plt.subplots()
  plt.axvline(x=0, color='lightgray')
  plt.axhline(y=0, color='lightgray')
  if (xlimits == None and ylimits == None):
    plt.xlim(np.min(x), np.max(x))
    plt.ylim(np.min(y), np.max(y))
  if (xlimits):
    plt.xlim(xlimits[0], xlimits[1])
  if (ylimits):
    plt.ylim(ylimits[0], ylimits[1])
  _ = ax.plot(x, y)


# plot multiple graphs
def plot_graphs(graphs):

  fig, ax = plt.subplots()

  for graph in graphs:
    ax.plot(graph[0], graph[1])

  plt.axvline(x=0, color='lightgray')
  plt.axhline(y=0, color='lightgray')


def plot_points(points, color=None, size=None, annotate='on'):
  for point in points: 
    plt.scatter(point[0], point[1], c=color, s=size)
    if (annotate == 'on'):
      plt.annotate(str(point), xy=(point[0]+0.25, point[1]+0.25), ha='center', va='top', c='black', fontsize=13)
    elif (annotate == 'off'):
      pass
````

## 4.1 Product Rule - for multiplicative comibinations

$$\frac{\delta (wz)}{\delta x} = w\frac{\delta (z)}{\delta x} +  z\frac{\delta (w)}{\delta x}$$

Useful:
- During calculations involving several rules
- Whenever product is incalculable pre-diff


Example) &emsp; $y =(6x^3)(7x^

4)$

$w=6x^3$ &emsp;&emsp;&emsp;&emsp;  $z=7x^4$

$\frac{\delta (w)}{\delta x} = 18x^2$ &emsp;&emsp; $\frac{\delta (z)}{\delta x} =28x^3$

$$\frac{\delta y}{\delta x} = \frac{\delta (wz)}{\delta x} = w\frac{\delta z}{\delta x} + z\frac{\delta w}{\delta x}$$
$$=(6x^3)(28x^3)+(7x^4)(18x^2)$$
$$=168x^6 + 126x^6$$
$$=294x^6$$

````python
def func41(x):
  y = 6*x**3 * 7*x**4
  return y

def deriv41(x):
  y = 294*x**6
  return y

x = np.linspace(-10, 10, 10000)
y41 = func41(x)
y41_der = deriv41(x)

plot_graphs([(x, y41), (x, y41_der)])
plt.annotate(f'$294x^6$', xy=(-8.5, deriv41(-8.5)), ha='center', va='top', c='navy', fontsize=15)
plt.annotate(f'$6x^3 * 7x^4$', xy=(-8.5, func41(-8.5)), ha='center', va='top', c='midnightblue', fontsize=15)
````

## 4.2 Quotient Rule - for fractions

$$\frac{\delta}{\delta x}(\frac{w}{z}) = \frac{z\frac{\delta (w)}{\delta x} +  w\frac{\delta (z)}{\delta x}}{z^2}$$

<br/>

E.g.) 
$$y=\frac{4x^2}{x^3+1}$$  
$$w=4x^2, \frac{dw}{dx}=8x$$  
$$z=x^3+1, \frac{dz}{dx}=3x^2$$  

<br />

$$\frac{\delta}{\delta x}(y) = \frac{\delta}{\delta x} (\frac{w}{z}) = \frac{(x^3+1)(8x) - (4x^2)(3X^2)}{(x^3+1)^2}$$  
$$\frac{8x^4 + 8x -12x^4}{(x^3+1)^2}$$  
$$\frac{-4x^4 + 8x}{(x^3+1)^2}$$  




````python
def func42(x):
  y = (4*x**2) / (x**3 +1)
  return y

def deriv42(x):
  y = (-4*x**4 + 8*x) /(x*3+1)**2
  return y

y42 = func42(x)
y42_der = deriv42(x)

plot_graphs([(x, y42), (x, y42_der)])
plt.xlim(-5, 5)
plt.ylim(-100, 100)
plt.annotate(f"$(4x^2)/(x^3+1)$", xy=(0.7, 75), ha='center', va='top', c='midnightblue', fontsize=15, fontweight='bold')
plt.annotate(f"$(-4x^4+8x)/(x^3+1)^2$", xy=(2.5, -25), ha='center', va='top', c='darkorange', fontsize=15, fontweight='bold')
````

## 4.3 Chain Rule - for nested functions
- Used for $\text{Gradient Descent}!$
- Many applications in ML
  - Critical for backpropagation algorithms used to train neutral nets
- Based on nested functions (a.k.a composite functions):
  - Let's say $y = (5x + 25)^3$
  - $ u = (5x + 25)^3$
  - $ y = u^3$
  - $y$ is a function of $u$, and $u$ is a function of $x$.
- *Chain rule* is an easy way to find derivative of nested function. 

````python
fig = plt.figure(figsize=(5.25, 0.75))
fig.text(0.5, 0.3, r"$\dfrac{\delta y}{\delta x} = \dfrac{\delta y}{\delta u} \dfrac{\delta u}{\delta x}$", 
         horizontalalignment='center', verticalalignment='center', fontsize=20, fontweight='bold')
plt.axis('off')
````

### Example $y=(2x^2 +8)^2$

chainrule ex.png

lecture slide from [4.3 Chain Rule](https://learning.oreilly.com/videos/calculus-for-machine/9780137398171/9780137398171-calm_01_04_03_00/) of the Calculus live lesson serise by Dr. Jon Krohn

````python
def func43(x):
  y = (2*x**2+8)**2
  return y

def deriv43(x):
  y = 16*x**3 + 64*x
  return y

y43 = func43(x)
y43_deriv = deriv43(x)

plot_graphs([(x, y43), (x, y43_deriv)])
plt.annotate(f"$(2x^2+8)^2$", xy=(-7.5, 20000), ha='left', va='center', fontsize=15, fontweight='bold')
plt.annotate(f"$16x^3+64x$", xy=(5, -5000), ha='center', va='bottom', fontsize=15, fontweight='bold')
````
